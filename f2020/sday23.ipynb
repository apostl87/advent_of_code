```python
# https://github.com/tmo1/adventofcode/blob/main/2020/23c.py

cups = {}
NUM = 1000000
start = [9,5,2,4,3,8,7,1,6]
#start = [3,8,9,5,4,7,6,1,2] + [i for i in range(10, NUM + 1)]
start = start + [i for i in range(10, NUM + 1)]
for i in range(len(start) - 1):
	cups[start[i]] = start[(i + 1)]
cups[start[-1]] = start[0]
current = start[0]

for i in range(10_000_000):
	if i % 100000 == 0:
		print(i)
    
	one = cups[current]
	two = cups[one]
	three = cups[two]
	cups[current] = cups[three]
	dest = current - 1
	while dest in [one, two, three] or dest == 0:
		if dest == 0:
			dest = NUM
		else:
			dest -= 1
	x = cups[dest]
	cups[dest] = one
	cups[three] = x
	current = cups[current]

print(cups[1] * cups[cups[1]])
```
