```python
from itertools import cycle
import numpy as np
from aoclib import list2string

lines = open('12.txt').read().strip()
```

```python
directions = {'E': np.array([1, 0]), 'S': np.array([0, -1]), 'W': np.array([-1, 0]), 'N': np.array([0, 1])}
turnright = {'E': 'S', 'S': 'W', 'W': 'N', 'N': 'E'}

facing, position = 'E', np.array([0, 0])
for instruction in lines.splitlines():
    action = instruction[0]
    number = int(instruction[1:])
    match action:
        case action if action in ['N', 'S', 'E', 'W']:
            position += directions[action]*number
        case 'L':
            for i in range((360-number)//90):
                facing = turnright[facing]
        case 'R':
            for i in range(number//90):
                facing = turnright[facing]
        case 'F':
            position += directions[facing]*number
print(facing, position)
print(sum([abs(x) for x in position]))
```

```python
directions = {'E': np.array([1, 0]), 'S': np.array([0, -1]), 'W': np.array([-1, 0]), 'N': np.array([0, 1])}

ship, waypoint = np.array([0, 0]), np.array((10, 1))
for instruction in lines.splitlines():
    action = instruction[0]
    number = int(instruction[1:])
    match action:
        case action if action in ['N', 'S', 'E', 'W']:
            waypoint += directions[action]*number
        case 'L':
            for i in range((360-number)//90):
                waypoint[0], waypoint[1] = waypoint[1], -waypoint[0]
        case 'R':
            for i in range(number//90):
                waypoint[0], waypoint[1] = waypoint[1], -waypoint[0]
        case 'F':
            ship += waypoint*number
print(ship, waypoint)
print(sum([abs(x) for x in ship]))
```
