```python
import re, sys, copy, numpy as np, copy
from math import sqrt, lcm
from aoclib import list2string
from collections import defaultdict
from itertools import combinations
from operator import add, mul
```

```python
inp = """
-1,2,2,0
0,0,2,-2
0,0,0,-2
-1,2,0,0
-2,-2,-2,2
3,0,2,-1
-1,3,2,2
-1,0,-1,0
0,2,1,-2
3,0,0,0
""".strip()

inp = open("25.txt").read().strip()
```

```python
def manhattan_distance(x, y):
    assert len(x) == len(y)
    return sum([abs(x[i] - y[i]) for i in range(len(x))])

# Initialize
pool = [tuple([int(x) for x in point.split(",")]) for i, point in enumerate(inp.splitlines())]
N = len(pool)

# Calculate distance matrix
D = np.full((N, N), np.infty)
for i in range(N):
    for j in range(i + 1, N):
        D[i, j] = D[j, i] = manhattan_distance(pool[i], pool[j])

# Build constellations
constellations = []
available = set(range(N))
while available:
    constellation = set()
    new = min(available)
    to_add = set({new})
    while to_add:
        new = to_add.pop()
        available.remove(new)
        constellation.add(new)
        to_add |= set([j for j in range(N) if D[new, j] <= 3 and j in available])
    constellations.append(frozenset(constellation))

len(constellations)
```
