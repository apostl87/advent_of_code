```python
import re, json
from collections import defaultdict
from itertools import cycle, combinations

input = open("3.input").read().strip()
lines = input.splitlines()
```

```python
grid = defaultdict(int)
not_overwritten = set()

regex = r'(\d+)'
for line in lines:
    n = re.findall(regex, line)
    id, x, y, nx, ny = [int(x) for x in n]
    not_overwritten.add(id)
    for i in range(nx):
        for j in range(ny):
            if grid[(x + i, y + j)] != 0:
                not_overwritten -= {grid[(x + i, y + j)], id}
                grid[(x + i, y + j)] = -1 
            else:
                grid[(x + i, y + j)] = id

print("Part 1:", len([x for x in grid.values() if x == -1]))
print("Part 2:", not_overwritten)
```
