```python
grid = input_.splitlines()
for i in range(len(grid)):
    grid[i] = list(grid[i])
m, n = len(grid), len(grid[0])

# substep function
def do_substep(grid, direction):
    changed = False
    new_grid = [['.' for i in range(n)] for j in range(m)]
    #print(list2string(grid))

    for y in range(m):
        for x in range(n):
            if grid[y][x] == direction:
                if direction == '>':
                    new_position = (y, (x + 1)%n)
                elif direction == 'v':
                    new_position = ((y + 1)%m, x)
                #print(grid[new_position[0]][new_position[1]])
                if grid[new_position[0]][new_position[1]] == '.':
                    changed = True
                    new_grid[new_position[0]][new_position[1]] = direction
                else:
                    new_grid[y][x] = direction
            elif grid[y][x] != '.':
                new_grid[y][x] = grid[y][x]

    #print(list2string(grid))
    return new_grid, changed

it = 0
while True:
    it += 1
    grid, ch1 = do_substep(grid, '>')
    grid, ch2 = do_substep(grid, 'v')
    if not ch1 and not ch2:
        print(ch1, ch2)
        break

#print_(cucumbers)
print("after", it, "iterations")
```
