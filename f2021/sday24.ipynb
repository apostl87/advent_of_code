```python
class ALU(object):
    def __init__(self):
        self.w, self.x, self.y, self.z = 0, 0, 0, 0
```

```python
w, x, y, z = 0, 0, 0, 0
def MONAD(inp, instructions):
    values = defaultdict(int)
    for i, instruction in enumerate(instructions):
        ins = instruction.split()
        if len(ins) > 2:
            try: y = int(ins[2])
            except: y = values[ins[2]]
        match ins[0]:
            case 'inp':
                values[ins[1]] = inp.pop(0)
            case 'add':
                values[ins[1]] += y
            case 'mul':
                values[ins[1]] *= y
            case 'div':
                if y == 0:
                    return False
                values[ins[1]] //= y
            case 'mod':
                if values[ins[1]] < 0 or y <= 0:
                    return False
                values[ins[1]] %= y
            case 'eql':
                values[ins[1]] = 1 if values[ins[1]] == y else 0
        if (i+1)%18 == 0:
            #print(values['x'], "\t", values['y'], "\t", values['z'])
            print(values['z'])
    return values
```
